```python
def Mol(V):
    k=1.380649*10**(-23)
    T=1000.0
    m=2.6566962*10**(-26)
    return float(4.0*math.pi*math.sqrt((m/(2.0*math.pi*k*T))**3)*(V**2)*math.exp((-m*V*V)/(2.0*k*T)))
```
