```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Гамма-распределение

https://ru.wikipedia.org/wiki/%D0%93%D0%B0%D0%BC%D0%BC%D0%B0-%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5

Первый параметр вводится через k, и равен 10 в моем случае

Второй считается равным 1, априори

Формулу среднего и дисперсии можно найти на вики

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html#scipy.stats.gamma

```python
gamma = sts.gamma
k=10.0
data = gamma.rvs(k, size=1000)
```

Само Распределение:

```python
plt.hist(data, bins = 30)
plt.show()
```

# Далее пойдут три гистограмы cреднего выборочного по 3, 10 и 50
параметры нормального распределения:

mu = k

sigma = sqrt(k/m)

```python
meannum = np.zeros(1000)
i=0
n=3
while i<1000:
    meannum[i]=np.mean(gamma.rvs(k, size=n))
    i = i+1
x = np.linspace(4,16,100)
norm_rv = sts.norm(k, math.sqrt(k/n))
pdf = norm_rv.pdf(x)
```

```python
plt.hist(meannum, bins = 30)
plt.plot(x, 450*pdf)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
meannum = np.zeros(1000)
i=0
n=10
while i<1000:
    meannum[i]=np.mean(gamma.rvs(k, size=n))
    i = i+1
x = np.linspace(6,14,100)
norm_rv = sts.norm(k, math.sqrt(k/n))
pdf = norm_rv.pdf(x)
```

```python
plt.hist(meannum, bins = 30)
plt.plot(x, 250*pdf)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
meannum = np.zeros(1000)
i=0
n=50
while i<1000:
    meannum[i]=np.mean(gamma.rvs(k, size=n))
    i = i+1
x = np.linspace(8,12,100)
norm_rv = sts.norm(k, math.sqrt(k/n))
pdf = norm_rv.pdf(x)
```

```python
plt.hist(meannum, bins = 20)
plt.plot(x, 120*pdf)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

# Вывод
При n=3 можно заметить некую скошенность гистограммы, что исправляется увелечение выборки

С повышением n гистограма становится более узкой, более сконцентрированной у среднего значения
