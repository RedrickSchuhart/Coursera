```python
def Func(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5.0 * np.exp(-x / 2.0)
```

```python
def PFunc(w1, w2, w3, w4, x):
    return w1+w2*x+w3*x*x+w4*x*x*x
```

```python
def newFunc(x):
    return 4.36264154+-1.29552587*x+0.19333685*x*x+-0.00823565*x*x*x
```

```python
A1 = np.zeros((2,2))
A3 = np.zeros((4,4))
f = np.zeros((4,1))
x = [1.0,4.0,10.0,15]
```

```python
i=0
while i<4:
    f[i,0]=Func(x[i])
    A3[i,0]=1
    A3[i,1]=x[i]
    A3[i,2]=x[i]*x[i]
    A3[i,3]=x[i]*x[i]*x[i]
    i+=1
```
